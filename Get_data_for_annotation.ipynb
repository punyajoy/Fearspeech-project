```python
selected_dataset={}

for key in valid_dataset:
    element=valid_dataset[key]
    count_dict={'normal':0,'fearspeech':0,'hatespeech':0}
    for annotation in element['annotations']:
        for label in annotation['Class']:
            if(label!='confused'):
                count_dict[label]+=1
    
    count_agree=0
    for label in count_dict.keys():
        if count_dict[label]==3:
            count_agree+=1
            
            
    for label in count_dict.keys():        
        if(label!='normal' and count_dict[label]==3):
            selected_dataset[key]={'id':element['id'],'text':element['text'],'label':label}
```

```python
###likert_df
final_dictionary={}
for key in dictionary_answers.keys():
    element=dictionary_answers[key]
    
    for i in range(1,3):
        for metric in ['belief','sharing']:
            temp1={'id':element['id']['post'+str(i)]+metric,
                   'label':element['label']['post'+str(i)],
                   'metric':metric,
                   'score':element[metric]['post'+str(i)]
                  }
            
            final_dictionary[temp1['id']]=temp1
    
     
df_likert=pd.DataFrame(final_dictionary).transpose().reset_index(drop=True)
df_likert
```

```python
belief_dict={'fearspeech':0,'hatespeech':0}
sharing_dict={'fearspeech':0,'hatespeech':0}
for key in dictionary_answers.keys():
    element=dictionary_answers[key]
    if(element['belief']['post1']>element['belief']['post2']):
        belief_dict[element['label']['post1']]+=1
    elif(element['belief']['post1']<element['belief']['post2']):
        belief_dict[element['label']['post2']]+=1
        
    if(element['sharing']['post1']>element['sharing']['post2']):
        sharing_dict[element['label']['post1']]+=1
    elif(element['sharing']['post1']<element['sharing']['post2']):
        sharing_dict[element['label']['post2']]+=1
```
